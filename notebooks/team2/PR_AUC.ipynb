```python
from pathlib import Path
import pandas as pd
from sklearn.metrics import average_precision_score, precision_recall_curve, auc

VALID_PREDICTIONS_PATH = Path(
    # "/content/drive/MyDrive/ieee_cis_fraud/member3_baseline/member3_lgbm_validation_predictions.parquet"
    "/content/member3_lgbm_validation_predictions.parquet"
)

valid_pred = pd.read_parquet(VALID_PREDICTIONS_PATH)

y_true = valid_pred["isFraud"]
y_score = valid_pred["prediction"]

# Common definition of PR AUC for imbalanced classification
pr_auc_ap = average_precision_score(y_true, y_score)

# Trapezoidal area under precision-recall curve
precision, recall, _ = precision_recall_curve(y_true, y_score)
pr_auc_trapz = auc(recall, precision)

print("PR AUC / Average Precision:", round(pr_auc_ap, 6))
print("PR AUC / trapezoidal:", round(pr_auc_trapz, 6))
```
